## -- Introducción --

**Objetivo:** Aplicar técnicas de preprocesamiento de texto para limpiar y preparar datos no estructurados antes de su análisis.  
**Dataset:** Opiniones simuladas en español sobre productos.  

En esta actividad se abordarán los pasos fundamentales del flujo de trabajo en minería de texto:  
1. Conversión a minúsculas.  
2. Eliminación de signos de puntuación.  
3. Tokenización.  
4. Eliminación de palabras vacías (stopwords).  
5. Lematización o stemming.  

El resultado será un texto limpio y listo para ser utilizado en modelos de Machine Learning.


## Importación de librerías y descarga de recursos

```python

import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Descargar recursos necesarios
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

##  Texto de ejemplo y preprocesamiento

```python
# Texto de ejemplo
texto = "¡La minería de texto es fascinante! Permite analizar grandes volúmenes de datos."

# Conversión a minúsculas y eliminación de puntuación
texto = texto.lower().translate(str.maketrans("", "", string.punctuation))

# Tokenización
tokens = word_tokenize(texto)

# Eliminación de palabras vacías
tokens_limpios = [word for word in tokens if word not in stopwords.words("spanish")]

print("Texto procesado:", tokens_limpios)
```

### Explicación del Código
- Se convierte el texto a minúsculas para evitar inconsistencias.  
- Se eliminan signos de puntuación, ya que no aportan significado semántico.  
- Se realiza la tokenización, dividiendo el texto en palabras individuales.  
- Se eliminan las palabras vacías (stopwords), como “de”, “es”, “la”, etc.  

**Ejemplo de salida:**  
`['minería', 'texto', 'fascinante', 'permite', 'analizar', 'grandes', 'volúmenes', 'datos']`


## -- Conclusión --

Después del preprocesamiento, el texto se encuentra limpio y estructurado, listo para ser utilizado en modelos de Machine Learning.  
Este paso es esencial para reducir ruido y mejorar la calidad de las representaciones numéricas del texto en etapas posteriores.
