## -- Introducción --

**Objetivo:** Comprender y aplicar las principales técnicas de representación de texto en Machine Learning.  
**Dataset:** Frases simuladas en español sobre minería de datos.  

En esta actividad se abordarán dos métodos fundamentales para transformar texto en vectores numéricos:  
1. **Bolsa de Palabras (Bag of Words – BoW)**  
2. **TF‑IDF (Term Frequency – Inverse Document Frequency)**  

Estas representaciones permiten que los algoritmos de aprendizaje automático procesen texto como datos estructurados.


## Bolsa de Palabras (BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer

# Documentos de ejemplo
documentos = [
    "La minería de texto es fascinante",
    "El análisis de datos es importante",
    "La minería de datos permite descubrir patrones"
]

# Crear el vectorizador BoW
vectorizador_bow = CountVectorizer()
bow_matriz = vectorizador_bow.fit_transform(documentos)

# Mostrar resultados
print("Términos:", vectorizador_bow.get_feature_names_out())
print("Matriz BoW:\n", bow_matriz.toarray())
```

## Explicación del Código (BoW)

### Explicación del Código
- Se crea una matriz donde cada columna representa una palabra única del corpus.  
- Cada celda indica cuántas veces aparece esa palabra en el documento.  
- No se considera el orden ni el significado de las palabras.  

**Ejemplo de salida:**  
Términos: ['análisis', 'datos', 'descubrir', 'es', 'fascinante', 'importante', 'la', 'minería', 'patrones', 'permite', 'texto']  
Matriz BoW:  
Cada fila representa un documento y cada columna la frecuencia de una palabra.


## TF‑IDF (Term Frequency – Inverse Document Frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Aplicar TF-IDF
vectorizador_tfidf = TfidfVectorizer()
tfidf_matriz = vectorizador_tfidf.fit_transform(documentos)

# Mostrar resultados
print("Términos:", vectorizador_tfidf.get_feature_names_out())
print("Matriz TF-IDF:\n", tfidf_matriz.toarray())
```

### Explicación del Código TF‑IDF
- TF‑IDF penaliza las palabras muy frecuentes y da más peso a las que son específicas de cada documento.  
- Es una representación más informativa que BoW, ya que refleja la relevancia de los términos.  

**Ejemplo de salida:**  
Términos: ['análisis', 'datos', 'descubrir', 'es', 'fascinante', 'importante', 'la', 'minería', 'patrones', 'permite', 'texto']  
Matriz TF‑IDF:  
Los valores son ponderaciones entre 0 y 1 que indican la importancia relativa de cada palabra.


## -- Conclusión --

Las representaciones BoW y TF‑IDF son la base para convertir texto en datos numéricos.  
Mientras BoW cuenta ocurrencias, TF‑IDF evalúa la relevancia de cada término.  
Ambas técnicas son esenciales para alimentar modelos de clasificación o análisis de sentimientos en minería de texto.
