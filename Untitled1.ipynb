**Mineria de Datos_Semana_1**
1. Identificación de Tipos de Datos
En el dataset de empleados:                   
**ID:** Numérico (entero).     
**Nombre** → Categórico (texto).   
**Edad** → Numérico (entero).                      
**Nivel Educativo** → Ordinal (Primaria < Secundaria < Licenciado < Ingeniero < Doctorado).                
**Salario** → Numérico (float).           
**Categoría** → Categórico (Junior, Senior, Manager).                           
**Estado:** Categórico (Activo/Inactivo).

**Manejo:**

Numéricos → se usan directamente en cálculos y modelos.
Categóricos → requieren codificación (One-Hot o Label Encoding).
Ordinales → se codifican con un orden lógico.
Fechas (si existieran) → se transforman en componentes (año, mes, día).

**2. Limpieza de Datos**
Problemas detectados en el dataset:
Edad negativa (-28).
Salario faltante (None).
Salario extremo (200000).
Inconsistencias en categóricos (Activo vs ACTIVO, “Licenciado” vs “licenciado”).
**Estrategias:**
Corregir edad negativa con valor absoluto.
Imputar salario faltante con la mediana.
Detectar outliers en salario con IQR.
Normalizar valores categóricos (pasar todo a minúsculas o capitalizar).
**3. Transformación de Datos**
Nivel Educativo → Codificación ordinal.
Categoría → One-Hot Encoding.
Salario → Normalización (MinMaxScaler).
Nueva variable → Años_Hasta_Jubilación = 65 - Edad.
**4. Análisis Exploratorio (EDA)**
Estadísticas descriptivas: media, mediana, min, max.
Visualizaciones:
Histograma de edades.
Boxplot de salarios.
Scatterplot Edad vs Salario.
Interpretación:
Detectar distribución de edades.
Ver dispersión de salarios y outliers.
Relación entre edad y salario.



**Mineria de Datos_Semana_1**
1. Identificación de Tipos de Datos
En el dataset de empleados:                   
**ID:** Numérico (entero).     
**Nombre** → Categórico (texto).   
**Edad** → Numérico (entero).                      
**Nivel Educativo** → Ordinal (Primaria < Secundaria < Licenciado < Ingeniero < Doctorado).                
**Salario** → Numérico (float).           
**Categoría** → Categórico (Junior, Senior, Manager).                           
**Estado:** Categórico (Activo/Inactivo).

**Manejo:**

Numéricos → se usan directamente en cálculos y modelos.
Categóricos → requieren codificación (One-Hot o Label Encoding).
Ordinales → se codifican con un orden lógico.
Fechas (si existieran) → se transforman en componentes (año, mes, día).

**2. Limpieza de Datos**
Problemas detectados en el dataset:
Edad negativa (-28).
Salario faltante (None).
Salario extremo (200000).
Inconsistencias en categóricos (Activo vs ACTIVO, “Licenciado” vs “licenciado”).
**Estrategias:**
Corregir edad negativa con valor absoluto.
Imputar salario faltante con la mediana.
Detectar outliers en salario con IQR.
Normalizar valores categóricos (pasar todo a minúsculas o capitalizar).
**3. Transformación de Datos**
Nivel Educativo → Codificación ordinal.
Categoría → One-Hot Encoding.
Salario → Normalización (MinMaxScaler).
Nueva variable → Años_Hasta_Jubilación = 65 - Edad.
**4. Análisis Exploratorio (EDA)**
Estadísticas descriptivas: media, mediana, min, max.
Visualizaciones:
Histograma de edades.
Boxplot de salarios.
Scatterplot Edad vs Salario.
Interpretación:
Detectar distribución de edades.
Ver dispersión de salarios y outliers.
Relación entre edad y salario.



```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder, OneHotEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset
df = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Nombre": ["Juan", "María", "Pedro", "Ana", "Luis"],
    "Edad": [32, -28, 40, 35, 50],
    "Nivel_Educativo": ["Licenciado", "Ingeniero", "Doctorado", "Licenciado", "Doctorado"],
    "Salario": [50000, 60000, None, 55000, 200000],
    "Categoría": ["Junior", "Senior", "Senior", "Junior", "Manager"],
    "Estado": ["Activo", "ACTIVO", "Inactivo", "Activo", "Inactivo"]
})

# Limpieza
df["Edad"] = df["Edad"].abs()
df["Salario"].fillna(df["Salario"].median(), inplace=True)
Q1, Q3 = df["Salario"].quantile([0.25, 0.75])
IQR = Q3 - Q1
df = df[~((df["Salario"] < (Q1 - 1.5 * IQR)) | (df["Salario"] > (Q3 + 1.5 * IQR)))]
df["Estado"] = df["Estado"].str.capitalize()

# Transformación
encoder_ordinal = OrdinalEncoder(categories=[["Licenciado", "Ingeniero", "Doctorado"]])
df["Nivel_Cod"] = encoder_ordinal.fit_transform(df[["Nivel_Educativo"]])

encoder_onehot = OneHotEncoder(sparse_output=False, drop="first")
encoded_cat = encoder_onehot.fit_transform(df[["Categoría"]])
df_encoded = pd.DataFrame(encoded_cat, columns=encoder_onehot.get_feature_names_out(["Categoría"]))
df = pd.concat([df, df_encoded], axis=1)

scaler = MinMaxScaler()
df["Salario_Normalizado"] = scaler.fit_transform(df[["Salario"]])

df["Años_Hasta_Jubilación"] = 65 - df["Edad"]

print(df)

# EDA
print(df.describe())
print(df[["Edad", "Salario"]].corr())

sns.histplot(df["Edad"], bins=5, kde=True)
plt.show()

sns.boxplot(x=df["Salario"])
plt.show()

sns.scatterplot(x="Edad", y="Salario", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma de edades
plt.figure(figsize=(6,4))
sns.histplot(df["Edad"], bins=5, kde=True, color="skyblue")
plt.title("Distribución de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# Boxplot de salarios
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Salario"], color="lightgreen")
plt.title("Boxplot de Salarios")
plt.xlabel("Salario")
plt.show()

# Scatterplot Edad vs Salario
plt.figure(figsize=(6,4))
sns.scatterplot(x="Edad", y="Salario", data=df, hue="Categoría", style="Estado", palette="deep")
plt.title("Relación entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
plt.show()
```

 **Interpretación de los gráficos**
**Histograma de Edades**: muestra cómo se distribuyen las edades de los empleados. La curva KDE suaviza la tendencia.
**Boxplot de Salarios**: permite detectar valores atípicos (outliers). Si ves puntos aislados fuera del rango, son salarios extremos.
**Scatterplot Edad vs Salario**: muestra la relación entre edad y salario, diferenciando categorías laborales por color y estado (Activo/Inactivo) por estilo de marcador. Esto ayuda a ver patrones, por ejemplo si los Senior tienden a tener salarios más altos.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Matriz de correlación
corr = df[["Edad", "Salario", "Salario_Normalizado", "Años_Hasta_Jubilación"]].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor de correlaciones")
plt.show()
```

**Interpretación**: Cada diagonal muestra la distribución de una variable (histogramas), y fuera de la diagonal se ven scatterplots entre pares de variables. Esto ayuda a detectar patrones y posibles relaciones no lineales.
