# Análisis Exploratorio de Datos (EDA)

En este notebook se realiza un análisis exploratorio sobre un conjunto de datos de clientes de una tienda online.

El objetivo es comprender cómo se distribuyen variables como la edad, los ingresos y la cantidad de compras, y detectar posibles patrones o comportamientos relevantes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Crear dataset
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

df
```

## Exploración inicial

Se revisa la estructura del dataset para conocer los tipos de datos y obtener estadísticas básicas.

```python
df.info()
```

```python
df.describe()
```

## Distribución de Edad

```python
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edad")
plt.show()
```

La mayoría de los clientes se concentra entre los 30 y 45 años. No se observan valores extremos en la edad, lo que indica una distribución bastante equilibrada.

## Distribución de Ingresos

```python
sns.histplot(df["Ingresos"], bins=5, kde=True)
plt.title("Distribución de Ingresos")
plt.show()
```

Los ingresos presentan una distribución relativamente uniforme, aunque se observa una leve concentración en valores medios-altos.

## Detección de valores atípicos en Compras

```python
sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.show()
```

No se observan valores atípicos extremos en la cantidad de compras. Los datos se encuentran dentro de un rango esperado.

## Relación entre Ingresos y Compras

```python
sns.scatterplot(x=df["Ingresos"], y=df["Compras"], hue=df["Categoría"])
plt.title("Relación entre Ingresos y Compras")
plt.show()
```

Se observa una relación positiva entre los ingresos y la cantidad de compras. A medida que aumentan los ingresos, también tiende a incrementarse el número de compras realizadas.

Además, se pueden distinguir diferencias según la categoría de cliente.

## Conclusión general

A partir del análisis exploratorio realizado, se puede observar que los clientes presentan una distribución relativamente equilibrada en cuanto a edad e ingresos, sin la presencia de valores atípicos significativos.

Además, se identifica una tendencia positiva entre los ingresos y la cantidad de compras, lo que sugiere que los clientes con mayor poder adquisitivo tienden a realizar más compras.

Por otro lado, las visualizaciones permiten distinguir ciertos patrones según la categoría de cliente, lo que podría ser útil para segmentar estrategias de marketing.

En general, el análisis exploratorio permitió comprender mejor la estructura de los datos y detectar relaciones clave que pueden ser profundizadas en análisis posteriores.                                     Este análisis sirve como base para futuros modelos predictivos o segmentación de clientes.
