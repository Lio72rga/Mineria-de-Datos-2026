# Visualización Avanzada de Datos

En este notebook se aplican técnicas de visualización avanzada para analizar relaciones entre múltiples variables y obtener insights más profundos.

```python
sns.set(style="whitegrid")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})
```

## Pairplot (Relación entre múltiples variables)

```python
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()
```

El pairplot permite observar relaciones entre todas las variables numéricas. Se aprecia claramente que ingresos y compras tienen una tendencia creciente conjunta.

## Heatmap de correlaciones

```python
correlaciones = df.corr(numeric_only=True)

sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de correlación")
plt.show()
```

El mapa de correlación muestra que la relación más fuerte se da entre ingresos y compras, lo que indica que los clientes con mayores ingresos tienden a comprar más.

## Violin Plot (Compras por categoría)

```python
sns.violinplot(x="Categoría", y="Compras", data=df)
plt.title("Distribución de compras por categoría")
plt.show()
```

Se observan diferencias en la distribución de compras según la categoría. Algunas categorías presentan mayor variabilidad en la cantidad de compras.
