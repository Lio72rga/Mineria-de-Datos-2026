# Comparación entre Clasificación y Regresión

## Regresión Logística
- Predice probabilidades entre 0 y 1.
- Se utiliza para clasificación.
- La salida representa categorías.

## Regresión Lineal
- Predice valores numéricos continuos.
- La salida puede tomar cualquier valor real.
- Se utiliza para estimar cantidades.

Ambos modelos pertenecen al aprendizaje supervisado, pero resuelven problemas distintos.

```python

df = sns.load_dataset("titanic").dropna()
df.head()
```

```python

X_class = df[["age", "fare"]]
y_class = df["survived"]

X_tr_c, X_te_c, y_tr_c, y_te_c = train_test_split(
    X_class, y_class, test_size=0.2, random_state=42
)

clf = LogisticRegression(max_iter=1000)
clf.fit(X_tr_c, y_tr_c)

print("Precisión:", clf.score(X_te_c, y_te_c))
```

```python

X_reg = df[["age"]]
y_reg = df["fare"]

X_tr_r, X_te_r, y_tr_r, y_te_r = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

reg = LinearRegression()
reg.fit(X_tr_r, y_tr_r)

print("R²:", reg.score(X_te_r, y_te_r))
```
