## -- Introducción --

**Objetivo:** Implementar un modelo de red neuronal con TensorFlow para clasificar opiniones en texto como positivas o negativas.  
**Dataset:** Opiniones simuladas en español sobre productos.  

En esta actividad se abordarán los pasos fundamentales:  
1. Tokenización y vectorización del texto.  
2. Aplicación de padding para uniformar longitudes.  
3. Construcción de un modelo secuencial con capas Embedding y LSTM.  
4. Entrenamiento y evaluación del modelo.  

El resultado será un clasificador binario capaz de distinguir entre opiniones positivas y negativas.


## Importación de librerías

```python
# Instalar TensorFlow (solo necesario en entornos virtuales o locales)
!pip install tensorflow

# Importación de librerías
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

## Datos de ejemplo

```python
# Opiniones de ejemplo
textos = [
    "Me encanta este producto",
    "Es un desastre, muy malo",
    "Excelente calidad y buen precio",
    "No me gustó, fue horrible"
]

# Etiquetas: 1 = Positivo, 0 = Negativo
etiquetas = np.array([1, 0, 1, 0])
```

## Tokenización y Padding

```python
# Tokenización del texto
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(textos)
sequences = tokenizer.texts_to_sequences(textos)

# Padding para uniformar longitudes
padded_sequences = pad_sequences(sequences, maxlen=5, padding='post')

print("Secuencias tokenizadas y padded:\n", padded_sequences)
```

## Definición del Modelo

```python
# Definir el modelo de red neuronal
modelo = keras.Sequential([
    Embedding(input_dim=1000, output_dim=16, input_length=5),
    LSTM(16),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compilar el modelo
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Entrenamiento y Evaluación

```python
# Entrenar el modelo
historial = modelo.fit(padded_sequences, etiquetas, epochs=5, verbose=1)

# Evaluar el modelo
loss, accuracy = modelo.evaluate(padded_sequences, etiquetas, verbose=0)
print(f"Precisión del modelo: {accuracy:.2f}")
```

### Explicación del Código
- Se tokenizan las opiniones y se convierten en secuencias numéricas.  
- Se aplica padding para que todas las secuencias tengan la misma longitud.  
- El modelo incluye una capa Embedding para representar palabras, seguida de una LSTM para aprender patrones en secuencias.  
- Se entrena el modelo con clasificación binaria (positivo/negativo).  
- La precisión indica el desempeño del clasificador en los datos de ejemplo.
